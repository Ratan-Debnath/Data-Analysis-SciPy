```
##help from geeks for geeks and scipy.org
#import numpy library
import numpy as np
#import linalg function from SciPy [linalg=linear algebra]
from scipy import linalg
a = np.array([[1,3,5],[2,4,6],[2,3,1]])
print(linalg.det(a)) #Compute determinant of a matrix
a #print
```

#Compute pivoted LU decomposition of a matrix
The decomposition is: 
A = P L U

where P is a permutation matrix, L lower triangular with unit diagonal elements, and U upper triangular.

```
p, l, u = linalg.lu(a)
print(p)
print(l)
print(u)
#print LU decomposition
np.dot(l,u)
```

```
#Eigen values and eigen vectors of this matrix
eigen_values, eigen_vectors = linalg.eig(a)
print(eigen_values)
print(eigen_vectors)
```

```
#sloving system of linear equations 
v = np.array([[2],[3],[5]])
print(v)
s = linalg.solve(a,v)
print(s)
```

```
#how to construct a large sparse matrix
from scipy import sparse
# Row-based linked list sparse matrix
a = sparse.lil_matrix((1000,1000))
print(a)
# output: <1000x1000 sparse matrix of type ''
#     with 0 stored elements in LInked List format>


a[0, :100] = np.random.rand(100)
a[1,100:200] = a[0, :100]
a.setdiag(np.random.rand(1000))
print(a)
#output: <1000x1000 sparse matrix of type ''
 #   with 1199 stored elements in LInked List format>
```

```
#linear algebra for Sparse Matrices
from scipy.sparse import  linalg
# Convert this matrix to Compressed Sparse Row format.
a = a.tocsr()
b = np.random.rand(1000)
ans = linalg.spsolve(a,b)
print(ans) #array of 1000 size
```

Integration: SciPy has integration methods in scipy.integrate module.
 
integrating e^{-x^2} over the range of 0 and 1 with respect to dx.
We will first define the function f(x)=e^(-x^2) , this is done using a lambda expression and then use quad routine.

```
import scipy.integrate
f = lambda x:np.exp(-x**2)
i = scipy.integrate.quad(f, 0, 1)
print(i) #first number value of integral and second means value of probable error
```

Double Integrals: function is scipy.integrate.dblquad(f,a,b,g,h)
Where, f is funtction of integrated, a and b are the lower and upper limits of the x variable.Also, g and h are the lower and upper ranges of the y variable.

As an example, let us perform the double integral of x*y^2 over x range from 0 to 2 and y ranges from 0 to 1.

```
from scipy import integrate
f = lambda y, x: x*y**2
i = integrate.dblquad(f, 0, 2, lambda x: 0, lambda x: 1)
print(i)
```
